```python
from annotation import Transcript
import pandas as pd
import glob
```

```python
transcripts = Transcript.hash_gff('/scratch/research/references/chlamydomonas/5.3_chlamy_w_organelles_mt_minus/annotation/concatenated_GFF/final.strict.GFF3', index_label='ness_ID', quiet=True)
```

```python
shared_df = pd.read_csv("../polymorphism/polymorphism_all.csv")
mtLimited_df = pd.read_csv("../polymorphism/polymorphism_all.mtLimited.csv")
nonMT_df =  pd.read_csv("../polymorphism/polymorphism_all.NonMTLocus.csv")
```

```python
name_translator = {}
for l in open("../find_shared_genes/NameTranslation.txt").readlines()[1:]:
    info = l.strip().split()
    domain = info[0]
    common_name = info[1]
    if len(info) == 4:
        mtPlus_ness_id = info[2]
        mtMinus_ness_id = info[3]
    elif len(info) ==2:
        continue
    elif len(info) == 3 and domain == 'C':
        mtPlus_ness_id = info[2]
        mtMinus_ness_id = info[2]
    else:print(l)
    name_translator[common_name]= {'mtPlus_ness_id':mtPlus_ness_id, 'mtMinus_ness_id':mtMinus_ness_id}

for c in name_translator:
    print(c, name_translator[c])
```

```python
mtPlus_ness_ids = []
mtMinus_ness_ids = []
for i in shared_df.gene:
    mtPlus_ness_ids.append(name_translator[i]['mtPlus_ness_id'])
    mtMinus_ness_ids.append(name_translator[i]['mtMinus_ness_id'])
    print(i, name_translator[i]['mtPlus_ness_id'], name_translator[i]['mtMinus_ness_id'])

shared_df['mtPlus_ness_ids'] = pd.Series(mtPlus_ness_ids, index=shared_df.index)
shared_df['mtMinus_ness_ids'] = pd.Series(mtMinus_ness_ids, index=shared_df.index)
```

```python
mtPlus_CDSs = []
for ness_ID in shared_df.mtPlus_ness_ids:
    print(ness_ID)
    t = transcripts[ness_ID]
    CDSs = [[exon.start, exon.end] for exon in t.sorted_feats('CDS')]
    mtPlus_CDSs.append(CDSs)
shared_df['mtPlus_CDSs'] = pd.Series(mtPlus_CDSs, index=shared_df.index)
```

```python
mtMinus_CDSs = []
for ness_ID in shared_df.mtMinus_ness_ids:
    t = transcripts[ness_ID]
    CDSs = [[exon.start, exon.end] for exon in t.sorted_feats('CDS')]
    mtMinus_CDSs.append(CDSs)
shared_df['mtMinus_CDSs'] = pd.Series(mtMinus_CDSs, index=shared_df.index)
```

```python
shared_df.to_csv('polymorphism.Shared.csv', index=False)
```
